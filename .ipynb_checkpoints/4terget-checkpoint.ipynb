```python
import torch 
from torch import nn

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

### data

```python
data = torch.randn((4000,2))
```

# linear => 1layer

```python
X = data
def func(x_1, x_2):
    if x_2>x_1:
        if x_2>-x_1:
            return [1,0,0,0]
        else:
            return [0,1,0,0]
    else:
        if x_2>-x_1:
            return [0,0,1,0]
        else:
            return [0,0,0,1]
y = []
for x_1, x_2 in zip(X[:,0], X[:,1]):
    y.append(func(x_1, x_2))
#add nois
X +=  torch.randn(4000,2)/4
```

```python
np.argmax(y, axis=1)
```

```python
x = np.arange(-5,5,0.01)
plt.plot(x ,x, label='y =x')
plt.plot(x ,-x, label='y =-x')
sns.scatterplot(X[:,0], X[:,1], hue=np.argmax(y, axis=1), alpha=0.3)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

```python
X = np.array(X)
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train = torch.from_numpy(X_train) .float() 
X_test = torch.from_numpy(X_test) .float() 
y_train = torch.from_numpy(y_train) .float() 
y_test = torch.from_numpy(y_test) .float() 
```

```python
model = torch.nn.Sequential(nn.Linear(in_features=2, out_features=4), 
                            nn.Softmax())
model = model.train()

mse = nn.MSELoss()
learning_rate = 0.02  
optimizer = torch.optim.SGD(model.parameters(),lr = learning_rate)
loss_list = []
loss_val_list = []
iteration_number =5001
for iteration in range(iteration_number):
        
    # optimization
    optimizer.zero_grad() 
    
    # Forward to get output
    results = model(X_train)
    # Calculate Loss
    loss = mse(results.squeeze(), y_train.squeeze())
    
    # backward propagation
    loss.backward()
    
    # Updating parameters
    optimizer.step()
    
    # store loss
    loss_list.append(loss.data)
    pred = model(X_test)
    loss_val = mse(pred.squeeze(), y_test.squeeze())
    loss_val_list.append(loss_val)
    # print loss
    if(iteration % 500 == 0):
        print('epoch {}, loss {}  val_loss {}'.format(iteration, loss.data, loss_val))
    
    
pred = model(X_test).squeeze().tolist()

df = pd.DataFrame(data={'x':X_test[:,0], 'y':X_test[:,1], 'y_test':np.argmax(y_test.squeeze(),axis=1), 'pred':np.argmax(pred, axis=1)})
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(data=df, x='x', y='y', hue='y_test')
plt.subplot(1,3,2)
sns.scatterplot(data=df, x='x', y='y', hue='pred')
plt.title(str((df.pred==df.y_test).mean())[:6])
plt.subplot(1,3,3)
plt.plot(loss_list, label='train')
plt.plot(loss_val_list, label='validation')
plt.legend()
plt.show()
```

# quadratic function => 2layer

```python
X = data
def func(x_1, x_2):
    if x_2>x_1**2-2:
        if x_2>2*x_1:
            return [1,0,0,0]
        else:
            return [0,1,0,0]
    else:
        if x_2>2*x_1:
            return [0,0,1,0]
        else:
            return [0,0,0,1]
        
y = []
for x_1, x_2 in zip(X[:,0], X[:,1]):
    y.append(func(x_1, x_2))
#add nois
X +=  torch.randn(4000,2)/4

x = np.arange(-5,5,0.01)
plt.plot(x ,x**2-2, label='y=x^2-1')
plt.plot(x, 2*x, label='y=2*x')
sns.scatterplot(X[:,0], X[:,1], hue=np.argmax(y,axis=1), alpha=0.3)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

```python
X = np.array(X)
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train = torch.from_numpy(X_train) .float() 
X_test = torch.from_numpy(X_test) .float() 
y_train = torch.from_numpy(y_train) .float() 
y_test = torch.from_numpy(y_test) .float() 

model = torch.nn.Sequential(nn.Linear(in_features=2, out_features=20), 
                            nn.ReLU(),
                            nn.Linear(20,  4),
                            nn.Softmax())
model = model.train()

mse = nn.MSELoss()
learning_rate = 0.02  
optimizer = torch.optim.SGD(model.parameters(),lr = learning_rate)
loss_list = []
loss_val_list = []
iteration_number =5001
for iteration in range(iteration_number):
        
    # optimization
    optimizer.zero_grad() 
    
    # Forward to get output
    results = model(X_train)
    # Calculate Loss
    loss = mse(results.squeeze(), y_train.squeeze())
    
    # backward propagation
    loss.backward()
    
    # Updating parameters
    optimizer.step()
    
    # store loss
    loss_list.append(loss.data)
    pred = model(X_test)
    loss_val = mse(pred.squeeze(), y_test.squeeze())
    loss_val_list.append(loss_val)
    # print loss
    if(iteration % 500 == 0):
        print('epoch {}, loss {}  val_loss {}'.format(iteration, loss.data, loss_val))
    
    
pred = model(X_test).squeeze().tolist()

df = pd.DataFrame(data={'x':X_test[:,0], 'y':X_test[:,1], 'y_test':np.argmax(y_test.squeeze(),axis=1), 'pred':np.argmax(pred, axis=1)})
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(data=df, x='x', y='y', hue='y_test')
plt.subplot(1,3,2)
sns.scatterplot(data=df, x='x', y='y', hue='pred')
plt.title(str((df.pred==df.y_test).mean())[:6])
plt.subplot(1,3,3)
plt.plot(loss_list, label='train')
plt.plot(loss_val_list, label='validation')
plt.legend()
plt.show()
```

# Cubic function =>  3layer

```python
X = data
def func(x_1, x_2):
    if x_2 > 0.5*(x_1**3) - 2*x_1 :
        if x_2 >-0.5*(x_1**3) + 2*x_1:
            return [1,0,0,0]
        else:
            return [0,1,0,0]
    else:
        if x_2 >-0.5*(x_1**3) + 2*x_1:
            return [0,0,1,0]
        else:
            return [0,0,0,1]
y = []
for x_1, x_2 in zip(X[:,0], X[:,1]):
    y.append(func(x_1, x_2))
#add nois
X +=  torch.randn(4000,2)/4

x = np.arange(-5,5,0.01)
plt.plot(x , 0.5*(x**3) - 2*x , label='y=x^3-4')
plt.plot(x , -0.5*(x**3) + 2*x , label='y=-x^3+4')

sns.scatterplot(X[:,0], X[:,1], hue=np.argmax(y,axis=1), alpha=0.3)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

```python
X = np.array(X)
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train = torch.from_numpy(X_train) .float() 
X_test = torch.from_numpy(X_test) .float() 
y_train = torch.from_numpy(y_train) .float() 
y_test = torch.from_numpy(y_test) .float() 

model = torch.nn.Sequential(nn.Linear(in_features=2, out_features=40), 
                            nn.ReLU(),
                            nn.Linear(40,  40),
                            nn.ReLU(),
                            nn.Linear(40,  4),
                            nn.Softmax())
model = model.train()

mse = nn.MSELoss()
learning_rate = 1e-4
optimizer = torch.optim.Adam(model.parameters(),lr = learning_rate)
loss_list = []
loss_val_list = []
iteration_number =5001
for iteration in range(iteration_number):
        
    # optimization
    optimizer.zero_grad() 
    
    # Forward to get output
    results = model(X_train)
    # Calculate Loss
    loss = mse(results.squeeze(), y_train.squeeze())
    
    # backward propagation
    loss.backward()
    
    # Updating parameters
    optimizer.step()
    
    # store loss
    loss_list.append(loss.data)
    pred = model(X_test)
    loss_val = mse(pred.squeeze(), y_test.squeeze())
    loss_val_list.append(loss_val)
    # print loss
    if(iteration % 500 == 0):
        print('epoch {}, loss {}  val_loss {}'.format(iteration, loss.data, loss_val))
    
    
pred = model(X_test).squeeze().tolist()

df = pd.DataFrame(data={'x':X_test[:,0], 'y':X_test[:,1], 'y_test':np.argmax(y_test.squeeze(),axis=1), 'pred':np.argmax(pred, axis=1)})
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(data=df, x='x', y='y', hue='y_test')
plt.subplot(1,3,2)
sns.scatterplot(data=df, x='x', y='y', hue='pred')
plt.title(str((df.pred==df.y_test).mean())[:6])
plt.subplot(1,3,3)
plt.plot(loss_list, label='train')
plt.plot(loss_val_list, label='validation')
plt.legend()
plt.show()
```

# 4, 5  layer

```python
X = data
def func(x_1, x_2):
    if x_2 > x_1**4 - 3*x_1**2:
        if x_2 >-0.5*(x_1**3) + 2*x_1:
            return [1,0,0,0]
        else:
            return [0,1,0,0]
    else:
        if x_2 >-0.5*(x_1**3) + 2*x_1:
            return [0,0,1,0]
        else:
            return [0,0,0,1]
y = []
for x_1, x_2 in zip(X[:,0], X[:,1]):
    y.append(func(x_1, x_2))
#add nois
X +=  torch.randn(4000,2)/4

x = np.arange(-5,5,0.01)
plt.plot(x , x**4 - 3*x**2)
plt.plot(x , -0.5*(x**3) + 2*x)
sns.scatterplot(X[:,0], X[:,1], hue=np.argmax(y,axis=1), alpha=0.3)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()

X = np.array(X)
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train = torch.from_numpy(X_train) .float() 
X_test = torch.from_numpy(X_test) .float() 
y_train = torch.from_numpy(y_train) .float() 
y_test = torch.from_numpy(y_test) .float() 

model = torch.nn.Sequential(nn.Linear(in_features=2, out_features=50), 
                            nn.ReLU(),
                            nn.Linear(50,  50),
                            nn.ReLU(),
                            nn.Linear(50,  50),
                            nn.ReLU(),
                            nn.Linear(50,  50),
                            nn.ReLU(),
                            nn.Linear(50,  4),
                            nn.Softmax())
model = model.train()

mse = nn.MSELoss()
learning_rate = 1e-3  
optimizer = torch.optim.Adam(model.parameters(),lr = learning_rate)
loss_list = []
loss_val_list = []
iteration_number =5001
for iteration in range(iteration_number):
        
    # optimization
    optimizer.zero_grad() 
    
    # Forward to get output
    results = model(X_train)
    # Calculate Loss
    loss = mse(results.squeeze(), y_train.squeeze())
    
    # backward propagation
    loss.backward()
    
    # Updating parameters
    optimizer.step()
    
    # store loss
    loss_list.append(loss.data)
    pred = model(X_test)
    loss_val = mse(pred.squeeze(), y_test.squeeze())
    loss_val_list.append(loss_val)
    # print loss
    if(iteration % 500 == 0):
        print('epoch {}, loss {}  val_loss {}'.format(iteration, loss.data, loss_val))
    
    
pred = model(X_test).squeeze().tolist()

df = pd.DataFrame(data={'x':X_test[:,0], 'y':X_test[:,1], 'y_test':np.argmax(y_test.squeeze(),axis=1), 'pred':np.argmax(pred, axis=1)})
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(data=df, x='x', y='y', hue='y_test')
plt.subplot(1,3,2)
sns.scatterplot(data=df, x='x', y='y', hue='pred')
plt.title(str((df.pred==df.y_test).mean())[:6])
plt.subplot(1,3,3)
plt.plot(loss_list, label='train')
plt.plot(loss_val_list, label='validation')
plt.legend()
plt.show()
```

```python
X = data
def func(x_1, x_2):
    if x_2 > 0.2*(x_1**5) - x_1**3 + 2*x_1:
        if x_2 >-(x_1**3) + 2*x_1:
            return [1,0,0,0]
        else:
            return [0,1,0,0]
    else:
        if x_2 >-(x_1**3) + 2*x_1:
            return [0,0,1,0]
        else:
            return [0,0,0,1]
y = []
for x_1, x_2 in zip(X[:,0], X[:,1]):
    y.append(func(x_1, x_2))
#add nois
X +=  torch.randn(4000,2)/4

x = np.arange(-5,5,0.01)
plt.plot(x , 0.2*(x**5)-x**3+2*x, label='f(x)')
plt.plot(x , -(x**3) + 2*x)
sns.scatterplot(X[:,0], X[:,1], hue=np.argmax(y,axis=1), alpha=0.3)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()

X = np.array(X)
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train = torch.from_numpy(X_train) .float() 
X_test = torch.from_numpy(X_test) .float() 
y_train = torch.from_numpy(y_train) .float() 
y_test = torch.from_numpy(y_test) .float() 

model = torch.nn.Sequential(nn.Linear(in_features=2, out_features=50), 
                            nn.ReLU(),
                            nn.Linear(40,  40),
                            nn.ReLU(),
                            nn.Linear(40,  40),
                            nn.ReLU(),
                            nn.Linear(40,  40),
                            nn.ReLU(),
                            nn.Linear(40,  40),
                            nn.ReLU(),
                            nn.Linear(40,  4),
                            nn.Sigmoid())
model = model.train()

mse = nn.MSELoss()
learning_rate = 0.02  
optimizer = torch.optim.SGD(model.parameters(),lr = learning_rate)
loss_list = []
loss_val_list = []
iteration_number =5001
for iteration in range(iteration_number):
        
    # optimization
    optimizer.zero_grad() 
    
    # Forward to get output
    results = model(X_train)
    # Calculate Loss
    loss = mse(results.squeeze(), y_train.squeeze())
    
    # backward propagation
    loss.backward()
    
    # Updating parameters
    optimizer.step()
    
    # store loss
    loss_list.append(loss.data)
    pred = model(X_test)
    loss_val = mse(pred.squeeze(), y_test.squeeze())
    loss_val_list.append(loss_val)
    # print loss
    if(iteration % 500 == 0):
        print('epoch {}, loss {}  val_loss {}'.format(iteration, loss.data, loss_val))
    
    
pred = model(X_test).squeeze().tolist()

df = pd.DataFrame(data={'x':X_test[:,0], 'y':X_test[:,1], 'y_test':np.argmax(y_test.squeeze(),axis=1), 'pred':np.argmax(pred, axis=1)})
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(data=df, x='x', y='y', hue='y_test')
plt.subplot(1,3,2)
sns.scatterplot(data=df, x='x', y='y', hue='pred')
plt.title(str((df.pred==df.y_test).mean())[:6])
plt.subplot(1,3,3)
plt.plot(loss_list, label='train')
plt.plot(loss_val_list, label='validation')
plt.legend()
plt.show()
```
